```python
# Q5  Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
# heatmap.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
n_samples = 100
x = np.random.rand(n_samples)  # Random feature
y = 2 * x + np.random.normal(0, 0.1, n_samples)  # Correlated feature with some noise
z = 0.5 * x + 0.5 * y + np.random.normal(0, 0.1, n_samples)  # Another correlated feature

# Create a DataFrame
data = pd.DataFrame({'Feature 1': x, 'Feature 2': y, 'Feature 3': z})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True, linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
"""
Q4 Using the given x and y data, generate a 3D surface plot to visualize the function x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
"""
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calculate z values
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = go.Figure(data=[go.Surface(z=df['Z'].values.reshape(100, 100),
                                   x=df['X'].values.reshape(100, 100),
                                   y=df['Y'].values.reshape(100, 100))])

# Update layout
fig.update_layout(title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))',
                  scene=dict(xaxis_title='X Axis',
                             yaxis_title='Y Axis',
                             zaxis_title='Z Axis'))

# Show the plot
fig.show()
```

```python
# Q2 Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
# markers based on the 'sizes' and 'colors' columns.
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# To display plots in the notebook (if you're using a Jupyter notebook)
output_notebook()

# Generate random data
N = 100  # Number of points
x = np.random.random(N) * 100  # Random x values
y = np.random.random(N) * 100  # Random y values

# Random sizes and colors
sizes = np.random.randint(5, 30, size=N)  # Sizes between 5 and 30
colors = np.random.choice(['blue', 'green', 'red', 'orange', 'purple'], size=N).tolist()  # Convert colors to a list of strings

# Create a data source
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a Bokeh figure
p = figure(title="Scatter Plot with Random Sizes and Colors",
           x_axis_label='X Values', y_axis_label='Y Values',
           width=700, height=500)  # Use 'width' and 'height' instead of 'plot_width' and 'plot_height'

# Add circles to the plot with variable sizes and colors
p.scatter(x='x', y='y', size='sizes', color='colors', source=source, alpha=0.6)

# Show the plot
show(p)
```

```python
"""
Q5 Create a Bokeh heatmap using the provided dataset.data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
"""
import numpy as np
from bokeh.io import show
from bokeh.plotting import figure
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.layouts import layout
from bokeh.palettes import Viridis256

# Data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data to work with Bokeh
x_flat = xx.flatten()
y_flat = yy.flatten()
values = data_heatmap.flatten()

# Create a color mapper (Viridis color palette)
mapper = LinearColorMapper(palette=Viridis256, low=min(values), high=max(values))

# Create a figure
p = figure(title="Heatmap", tools="hover", tooltips=[("Value", "@image")], x_range=(0, 1), y_range=(0, 1))

# Create the heatmap using image glyph
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, palette=Viridis256)

# Add a color bar for the heatmap
color_bar = ColorBar(color_mapper=mapper, label_standoff=12, location=(0, 0), title='Values')
p.add_layout(color_bar, 'right')

# Show the heatmap
show(p)
```
